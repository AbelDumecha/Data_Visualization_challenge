```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
from scipy.stats import linregress


# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single DataFrame
mouse_study_results = pd.merge(study_results, mouse_metadata, how="left", on=["Mouse ID", "Mouse ID"])
# Display the data table for preview
mouse_study_results.head()
```

```python
# Checking the number of mice.
number_of_mice = mouse_study_results["Mouse ID"].nunique()
number_of_mice
```

```python
# Our data should be uniquely identified by Mouse ID and Timepoint
# Get the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
duplicate_mice = mouse_study_results.loc[mouse_study_results.duplicated(subset=["Mouse ID", "Timepoint"], keep=False), "Mouse ID"].unique()
duplicate_mice
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
duplicate_mouse_data_df = mouse_study_results[mouse_study_results.duplicated(subset=["Mouse ID", "Timepoint"], keep=False)]
duplicate_mouse_id = duplicate_mouse_data['Mouse ID'].unique()
duplicate_mouse_data = mouse_study_results[mouse_study_results["Mouse ID"].isin(duplicate_mouse_id)]
duplicate_mouse_data
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
cleaned_mouse_data = mouse_study_results[mouse_study_results["Mouse ID"].isin(duplicate_mouse_id)==False]
#cleaned_mouse_data = mouse_study_results.drop_duplicates()
cleaned_mouse_data
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 

# mean, median, variance, standard deviation, and SEM of the tumor volume. 

mean = cleaned_mouse_data.groupby(['Drug Regimen'])['Tumor Volume (mm3)'].mean()
median = cleaned_mouse_data.groupby(['Drug Regimen'])['Tumor Volume (mm3)'].median()
variance = cleaned_mouse_data.groupby(['Drug Regimen'])['Tumor Volume (mm3)'].var()
standard_deviation = cleaned_mouse_data.groupby(['Drug Regimen'])['Tumor Volume (mm3)'].std()
SEM = cleaned_mouse_data.groupby(['Drug Regimen'])['Tumor Volume (mm3)'].sem()

# Assemble the resulting series into a single summary DataFrame.
statistics_summary_df = {
    "Mean Tumor Volume":mean, "Median Tumor Volume":median,"Tumor Volume Variance":variance,"Tumor Volume Std. Dev.":standard_deviation,
    "Tumor Volume Std. Err.":SEM
}

statistics_summary = pd.DataFrame(statistics_summary_df)

statistics_summary
    
```

```python
# A more advanced method to generate a summary statistics table of mean, median, variance, standard deviation,
# and SEM of the tumor volume for each regimen (only one method is required in the solution)

# Using the aggregation method, produce the same summary statistics in a single line
statistics_summary_opt = cleaned_mouse_data.groupby('Drug Regimen')['Tumor Volume (mm3)'].agg(['mean', 'median', 'var', 'std', 'sem'])
statistics_summary_opt
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin
selected_treatment = ['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin']

#filter from the DataFrame

filtered_selected_treatment = cleaned_mouse_data[cleaned_mouse_data['Drug Regimen'].isin(selected_treatment)]
tumor_min_volume = filtered_selected_treatment.groupby(['Mouse ID', 'Drug Regimen'])['Tumor Volume (mm3)'].min().reset_index()

# Start by getting the last (greatest) timepoint for each mouse

timepoint_max = cleaned_mouse_data.groupby(['Mouse ID'])['Timepoint'].max().reset_index()

# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
merged_mouse_study_data = pd.merge(tumor_min_volume, timepoint_max, how="left", on=['Mouse ID'])
merged_mouse_study_data
```

```python
# Put treatments into a list for for loop (and later for plot labels)

treatments_list = ['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin']

# Create empty list to fill with tumor vol data (for plotting)
tumor_vol_data = []

# Calculate the IQR and quantitatively determine if there are any potential outliers. 
# Locate the rows which contain mice on each drug and get the tumor volumes

    # add subset 

    # Determine outliers using upper and lower bounds

# loop through each treatments
for treatment in treatments_list:
    
    #filter the data for the corresponding current treatment
    treatment_data = merged_mouse_study_data[merged_mouse_study_data['Drug Regimen'] == treatment]

    #get the tumor volume data for the current treatment
    tumor_volumes = treatment_data['Tumor Volume (mm3)']
    
    #append the tumor volume data to the list
    tumor_vol_data.append(tumor_volumes)
    
    # Calculate the IQR (Interquartile Range) for the current treatment regimen
    quartiles = tumor_volumes.quantile([0.25, 0.5, 0.75])
    lower_q = quartiles[0.25]
    upper_q = quartiles[0.75]
    iqr = upper_q - lower_q

    # Determine the upper and lower bounds to identify potential outliers
    lower_bound = lower_q - (1.5 * iqr)
    upper_bound = upper_q + (1.5 * iqr)

    # Locate potential outliers for the current treatment regimen
    potential_outliers = treatment_data[(treatment_data['Tumor Volume (mm3)'] < lower_bound) | 
                                        (treatment_data['Tumor Volume (mm3)'] > upper_bound)]
    
    # Print or store the potential outliers for further analysis
    print(f"{treatment}'s potential outliers: {potential_outliers['Tumor Volume (mm3)']}")
```
